```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("nameContainer", "unit-catalog")
dbutils.widgets.text("nameStorage",   "adlsantodata1703")
dbutils.widgets.text("catalogo",      "catalog_au")
```

```sql
%python

nameContainer = dbutils.widgets.get("nameContainer")
nameStorage = dbutils.widgets.get("nameStorage")
catalogo = dbutils.widgets.get("catalogo")

ruta = f"abfss://{nameContainer}@{nameStorage}.dfs.core.windows.net"
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.spotify_tracks;
DROP TABLE IF EXISTS ${catalogo}.bronze.spotify_daily_charts;
```

```sql
%python
## REMOVE DATA (Bronze)
dbutils.fs.rm(f"{ruta}/bronze/spotify_tracks",        True)
dbutils.fs.rm(f"{ruta}/bronze/spotify_daily_charts",  True)
```

### Eliminacion tablas Silver

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.silver.spotify_transformed;
```

```sql
%python
# REMOVE DATA (Silver)
dbutils.fs.rm(f"{ruta}/silver/spotify_transformed", True)
```

### Eliminacion tablas Golden

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.golden.spotify_final_summary;
```

```sql
%python
# REMOVE DATA (Golden)
dbutils.fs.rm(f"{ruta}/golden/spotify_final_summary", True)
```

### Eliminacion de Schemas

```sql
-- DROP SCHEMAS
DROP SCHEMA IF EXISTS ${catalogo}.bronze     CASCADE;
DROP SCHEMA IF EXISTS ${catalogo}.silver     CASCADE;
DROP SCHEMA IF EXISTS ${catalogo}.golden     CASCADE;
DROP SCHEMA IF EXISTS ${catalogo}.raw        CASCADE;
DROP SCHEMA IF EXISTS ${catalogo}.exploratory CASCADE;
```

### Eliminacion del Catalogo

```sql
%python
## REMOVE DATA (Catalogo completo)
dbutils.fs.rm(ruta, True)
```
