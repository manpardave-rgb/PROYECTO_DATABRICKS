## Grants — Proyecto Spotify Medallion

```sql
-- Permitir uso del catalogo al usuario principal
GRANT USE CATALOG ON CATALOG catalog_au TO `loincognoscible@gmail.com`;

-- Permitir uso del catalogo al grupo de ingenieros
GRANT USE CATALOG ON CATALOG catalog_au TO `DEs`;
```

```sql
-- Bronze
GRANT USE SCHEMA ON SCHEMA catalog_au.bronze TO `loincognoscible@gmail.com`;
GRANT USE SCHEMA ON SCHEMA catalog_au.bronze TO `DEs`;
GRANT CREATE TABLE ON SCHEMA catalog_au.bronze TO `DEs`;

-- Silver
GRANT USE SCHEMA ON SCHEMA catalog_au.silver TO `loincognoscible@gmail.com`;
GRANT USE SCHEMA ON SCHEMA catalog_au.silver TO `DEs`;
GRANT CREATE TABLE ON SCHEMA catalog_au.silver TO `DEs`;

-- Golden
GRANT USE SCHEMA ON SCHEMA catalog_au.golden TO `loincognoscible@gmail.com`;
GRANT USE SCHEMA ON SCHEMA catalog_au.golden TO `DEs`;
GRANT CREATE TABLE ON SCHEMA catalog_au.golden TO `DEs`;
```

```sql
-- spotify_tracks: solo lectura para CTs, lectura y escritura para DEs
GRANT SELECT ON TABLE catalog_au.bronze.spotify_tracks TO `loincognoscible@gmail.com`;
GRANT SELECT ON TABLE catalog_au.bronze.spotify_tracks TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.spotify_tracks TO `CTs`;

-- spotify_daily_charts: solo lectura para CTs, lectura y escritura para DEs
GRANT SELECT ON TABLE catalog_au.bronze.spotify_daily_charts TO `loincognoscible@gmail.com`;
GRANT SELECT ON TABLE catalog_au.bronze.spotify_daily_charts TO `DEs`;
GRANT SELECT ON TABLE catalog_au.bronze.spotify_daily_charts TO `CTs`;
```

```sql
-- spotify_transformed: lectura para analitica, escritura para ingenieria
GRANT SELECT ON TABLE catalog_au.silver.spotify_transformed TO `loincognoscible@gmail.com`;
GRANT SELECT ON TABLE catalog_au.silver.spotify_transformed TO `DEs`;
GRANT SELECT ON TABLE catalog_au.silver.spotify_transformed TO `CTs`;
```

```sql
-- spotify_final_summary: lectura para todos, escritura solo para DEs
GRANT SELECT ON TABLE catalog_au.golden.spotify_final_summary TO `loincognoscible@gmail.com`;
GRANT SELECT ON TABLE catalog_au.golden.spotify_final_summary TO `DEs`;
GRANT SELECT ON TABLE catalog_au.golden.spotify_final_summary TO `CTs`;
```

```sql
-- Lectura del raw (donde estan los CSVs de Spotify)
GRANT READ FILES ON EXTERNAL LOCATION `exlt-raw` TO `DEs`;

-- Escritura en bronze
GRANT WRITE FILES ON EXTERNAL LOCATION `exlt-bronze` TO `DEs`;

-- Escritura en silver
GRANT WRITE FILES ON EXTERNAL LOCATION `exlt-silver` TO `DEs`;

-- Escritura en golden
GRANT WRITE FILES ON EXTERNAL LOCATION `exlt-golden` TO `DEs`;

-- Escritura en unit-catalog (managed storage del catalogo)
GRANT ALL PRIVILEGES ON EXTERNAL LOCATION `exlt-unit-catalog` TO `DEs`;
```
